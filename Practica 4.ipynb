```
import numpy as np
import sys

class CMatFloat:

    def __init__(self):
        self.Matriz = None
        self._m_nFilas = 0
        self._m_nColumnas = 0

    def CrearMatriz2D(self, nFilas, nColumnas):
        if nFilas > 0 and nColumnas > 0:
            self.Matriz = np.zeros((nFilas, nColumnas), dtype=float)
            self._m_nFilas = nFilas
            self._m_nColumnas = nColumnas
        else:
            print("ERROR: El número de filas y columnas debe ser positivo.")
            self.Matriz = None
            self._m_nFilas = 0
            self._m_nColumnas = 0

    def CrearMatriz1D(self, nElementos):
        self.CrearMatriz2D(1, nElementos)

    def Introducir(self):
        if not self.Existe():
            print("ERROR: La matriz no ha sido creada. Use las opciones 1 o 2.")
            return

        print("\n--- Introducir Elementos de la Matriz ---")
        for i in range(self._m_nFilas):
            for j in range(self._m_nColumnas):
                mensaje = f"Introduce el valor para la posición [{i+1}, {j+1}] (decimal): "
                valor = leer_float(mensaje)
                self.Matriz[i, j] = valor
        print("Elementos introducidos correctamente.")

    def Mostrar(self):
        if self.Existe():
            print("\n--- Matriz Actual ---")
            print(self.Matriz)
            print(f"Dimensiones: {self._m_nFilas} Filas x {self._m_nColumnas} Columnas")
        else:
            print("ERROR: La matriz no existe o está vacía.")

    def Existe(self):
        return self.Matriz is not None and self._m_nFilas > 0 and self._m_nColumnas > 0

    def SumarMatrices(self, otra_matriz):
        if not self.Existe() or not otra_matriz.Existe():
            print("ERROR: Una o ambas matrices no existen.")
            return None

        if self._m_nFilas != otra_matriz._m_nFilas or self._m_nColumnas != otra_matriz._m_nColumnas:
            print("ERROR: Las dimensiones de las matrices no coinciden para la suma.")
            print(f"Matriz 1: {self._m_nFilas}x{self._m_nColumnas}, Matriz 2: {otra_matriz._m_nFilas}x{otra_matriz._m_nColumnas}")
            return None

        resultado = self.Matriz + otra_matriz.Matriz
        return resultado

    def RestarMatrices(self, otra_matriz):
        if not self.Existe() or not otra_matriz.Existe():
            print("ERROR: Una o ambas matrices no existen.")
            return None

        if self._m_nFilas != otra_matriz._m_nFilas or self._m_nColumnas != otra_matriz._m_nColumnas:
            print("ERROR: Las dimensiones de las matrices no coinciden para la resta.")
            print(f"Matriz 1: {self._m_nFilas}x{self._m_nColumnas}, Matriz 2: {otra_matriz._m_nFilas}x{otra_matriz._m_nColumnas}")
            return None

        resultado = self.Matriz - otra_matriz.Matriz
        return resultado
```

```
def leer_int(mensaje="Introduce un número entero: "):
    while True:
        try:
            entrada = input(mensaje)
            return int(entrada)
        except ValueError:
            print("¡Error! La entrada no es un número entero válido. Inténtalo de nuevo.")

def leer_float(mensaje="Introduce un número decimal: "):
    while True:
        try:
            entrada = input(mensaje)
            entrada = entrada.replace(',', '.')
            return float(entrada)
        except ValueError:
            print("¡Error! La entrada no es un número decimal válido. Inténtalo de nuevo.")

def crear_menu(opciones_menu):
    print("\n" + "="*40)
    for i, opcion in enumerate(opciones_menu, 1):
        print(f"{i}. {opcion}")
    print("="*40)

    num_opciones = len(opciones_menu)
    while True:
        seleccion = leer_int("Selecciona una opción: ")

        if 1 <= seleccion <= num_opciones:
            return seleccion
        else:
            print(f"Opción no válida. Debe ser un número entre 1 y {num_opciones}.")
```

```
def sub_menu_operaciones(matriz_a):
    opciones_sub_menu = [
        "Sumar matrices",
        "Restar matrices",
        "Volver al menú principal"
    ]

    while True:
        print("\n--- SUBMENÚ: Operaciones con Matrices ---")
        opcion = crear_menu(opciones_sub_menu)

        if opcion == 1 or opcion == 2:
            print("\n--- Preparando Segunda Matriz para la Operación ---")

            matriz_b = CMatFloat()
            matriz_b.CrearMatriz2D(matriz_a._m_nFilas, matriz_a._m_nColumnas)

            print(f"Se creará una Matriz B de {matriz_a._m_nFilas}x{matriz_a._m_nColumnas}.")
            matriz_b.Introducir()

            if opcion == 1:
                resultado = matriz_a.SumarMatrices(matriz_b)
                if resultado is not None:
                    print("\n[ RESULTADO DE LA SUMA ]")
                    print(resultado)
            elif opcion == 2:
                resultado = matriz_a.RestarMatrices(matriz_b)
                if resultado is not None:
                    print("\n[ RESULTADO DE LA RESTA ]")
                    print(resultado)

        elif opcion == 3:
            break

def main():
    matriz_a = CMatFloat()

    opciones_menu_principal = [
        "Construir matriz 1D",
        "Construir matriz 2D",
        "Introducir matriz",
        "Mostrar matriz",
        "Operaciones con matrices",
        "Terminar"
    ]

    print("=== PROGRAMA DE GESTIÓN DE MATRICES (NUMPY) ===")

    while True:
        opcion = crear_menu(opciones_menu_principal)

        if opcion == 1:
            n_elementos = leer_int("Introduce el número de elementos del vector (1D): ")
            matriz_a.CrearMatriz1D(n_elementos)
            if matriz_a.Existe():
                print(f"Matriz 1D de {n_elementos} elementos creada (1 fila x {n_elementos} columnas).")

        elif opcion == 2:
            n_filas = leer_int("Introduce el número de filas: ")
            n_columnas = leer_int("Introduce el número de columnas: ")
            matriz_a.CrearMatriz2D(n_filas, n_columnas)
            if matriz_a.Existe():
                print(f"Matriz 2D de {n_filas}x{n_columnas} creada.")

        elif opcion == 3:
            matriz_a.Introducir()

        elif opcion == 4:
            matriz_a.Mostrar()

        elif opcion == 5:
            if matriz_a.Existe():
                sub_menu_operaciones(matriz_a)
            else:
                print("ERROR: Debes crear la primera matriz (opción 1 o 2) antes de realizar operaciones.")

        elif opcion == 6:
            print("\nPrograma finalizado. ¡Hasta luego!")
            sys.exit(0)

if __name__ == "__main__":
    main()
```
